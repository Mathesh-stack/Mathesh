```python
# Relationship between categorical and continuous variable
sns.boxplot(x="Density (per km²)", y="World Population Percentage", data = data)  # Age distribution of gym goers with respect to gender
```

```python
# Relationship between categorical and continuous variable
sns.boxplot(x="Growth Rate", y="World Population Percentage", data = data)  # Product distribution of gym goers with respect to gender
```

```python
#Relationship between two categorical variables
sns.countplot(x="Growth Rate", hue = "World Population Percentage", data = data)
```
